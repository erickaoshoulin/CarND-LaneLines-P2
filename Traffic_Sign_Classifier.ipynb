````python
import tensorflow as tf
from tensorflow.contrib.layers import flatten
print('TensorFlow Version: {}'.format(tf.__version__))

### Define your architecture here.
### Feel free to use as many code cells as needed.


def LeNet_old(x, keep_prob):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # SOLUTION: Layer 1: Convolutional. Input = 32x32x3. Output = 28x28x64.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 3, 24), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(24))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    # SOLUTION: Activation.
    conv1 = tf.nn.relu(conv1)

    # SOLUTION: Pooling. Input = 28x28x256. Output = 14x14x256.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    #add drop out
    #conv1 = tf.nn.dropout(conv1, keep_prob)


    # SOLUTION: Layer 2: Convolutional. input = 14x14x32. Output 10x10x32
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 24, 64), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(64))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
        
    # SOLUTION: Activation.
    conv2 = tf.nn.relu(conv2)

    # SOLUTION: Pooling. Input = 10x10x64. Output = 5x5x64.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    #add drop out
    #conv2 = tf.nn.dropout(conv2, keep_prob)

    # SOLUTION: Flatten. Input = 3x3x64. Output = 1600
    fc0   = flatten(conv2)
    
    # SOLUTION: Layer 3: Fully Connected. Input = 4608. Output = 512.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(1600, 128), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(128))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)
    fc1 = tf.nn.dropout(fc1, keep_prob)

    # SOLUTION: Layer 4: Fully Connected. Input = 128. Output = 96.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(128, 96), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(96))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    #logits = tf.matmul(fc1, fc2_W) + fc2_b
        
    # SOLUTION: Activation.
    fc2    = tf.nn.relu(fc2)
    
    #add drop out
    #fc2 = tf.nn.dropout(fc2, keep_prob)


    # SOLUTION: Layer 5: Fully Connected. Input = 96. Output = n_classes = 43.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(96, n_classes), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(n_classes))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits



layer_one_features = 32
layer_two_features = 64
fc1_features = 124
fc2_features = 96

def LeNet(x, keep_prob):    
    # Hyperparameters
    mu = 0
    sigma = 0.1
    
    # SOLUTION: Layer 1: Convolutional. Input = 32x32x3. Output = 28x28xlayer_one_features.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 3, layer_one_features), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(layer_one_features))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    # SOLUTION: Activation.
    conv1 = tf.nn.relu(conv1)

    # SOLUTION: Pooling. Input = 28x28xlayer_one_features. Output = 14x14xlayer_one_features.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # SOLUTION: Layer 2: Convolutional. Output = 10x10xlayer_two_features.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, layer_one_features, layer_two_features), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(layer_two_features))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    
    # SOLUTION: Activation.
    conv2 = tf.nn.relu(conv2)

    # SOLUTION: Pooling. Input = 10x10xlayer_two_features. Output = 5x5xlayer_two_features.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # SOLUTION: Flatten. Input = 5x5xlayer_two_features. Output = 5*5*layer_two_features(flat).
    fc0   = flatten(conv2)

    # SOLUTION: Layer 3: Fully Connected. Input = 5*5*layer_two_features. Output = fc1_features.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(5*5*layer_two_features, fc1_features), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(fc1_features))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)

    # SOLUTION: Layer 4: Fully Connected. Input = 120. Output = fc2_features.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(fc1_features, fc2_features), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(fc2_features))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # SOLUTION: Activation.
    fc2    = tf.nn.relu(fc2)

    # SOLUTION: Layer 5: Fully Connected. Input = fc2_features. Output = 43.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(fc2_features, 43), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
````
